```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

# Korean -> English Machine Translation
tokenizer = AutoTokenizer.from_pretrained("QuoQA-NLP/KE-T5-Ko2En-Base")
model = AutoModelForSeq2SeqLM.from_pretrained("QuoQA-NLP/KE-T5-Ko2En-Base")
```

```python

# https://huggingface.co/datasets/conceptual_captions
src_text = [
    "안녕하세요, 일러스트 그림 그려주세요. 후드티를 입고 고양이 귀를 한 분홍머리 여자, 나이는 16살 정도, 일본 애니메이션 스타일로 부탁드려요."
]



translated = model.generate(
    **tokenizer(src_text, return_tensors="pt", padding=True, max_length=64,),
    max_length=64,
    num_beams=3,
    repetition_penalty=0.3,
    no_repeat_ngram_size=2,
    num_return_sequences=1,
)


[tokenizer.decode(t, skip_special_tokens=True) for t in translated]
#tokenizer.decode(translated, skip_special_tokens=True)
```
